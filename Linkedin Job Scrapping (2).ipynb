```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

import os
import warnings
warnings.filterwarnings("ignore")
import requests
from bs4 import BeautifulSoup
```

## DATA ANALYST JOBS

```python
url_DataAnalysis = "https://www.linkedin.com/jobs/search/?currentJobId=3593352924&keywords=data%20analyst"
```

```python
locations = []
titles = []
countries = []    
links = []
company_name = []


for i in range(0,26,25):
    url = url_DataAnalysis+"&start="+str(i)
    request = requests.get(url)
    data = request.text
    soup = BeautifulSoup(data)
    location = soup.findAll("span", {"class":"job-search-card__location"})
    for i in location:
        locations.append(i.text.replace("\n"," ").strip(" "))   
    title = soup.findAll("h3",{"class":"base-search-card__title"})
    for i in title:
        titles.append(i.text.replace("\n"," ").strip(" ")) 
    country = soup.findAll("span", {"class":"job-search-card__location"})
    for i in country:
        countries.append(i.text.replace("\n"," ").strip(" ").split(',')[-1])
    job_links = soup.findAll("a",{"class":"base-card__full-link"})
    for i in job_links:
        links.append(i["href"])
        
    names = soup.findAll("a",{"class":"hidden-nested-link"})
    for i in names:
        company_name.append(i.text.replace("\n"," ").strip(" ")) 
```

```python
print(len(locations))
print(len(titles))
print(len(countries))
print(len(links))
print(len(company_name))
```

```python
df1 = pd.DataFrame(titles,columns=["Title"])
df1["Location"] = locations
df1["Country"] = countries
df1["Links"] = links
df1["Company_Name"] = company_name

df1
```
